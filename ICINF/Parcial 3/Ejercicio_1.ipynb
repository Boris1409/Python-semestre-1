```python
datos = datos.head(20)
datos.set_index('Position', inplace = True)

ax = datos['Goals For'].plot(kind='bar')

for i, rating in enumerate(datos['Goals For']):
    ax.annotate(str(rating), xy=(i, rating), ha='center', va='bottom')

plt.show()
```

```python
datos2 = datos2.head(20)
datos2.set_index('Position', inplace = True)

ax = datos2['Goals For'].plot(kind='bar')

for i, rating in enumerate(datos2['Goals For']):
    ax.annotate(str(rating), xy=(i, rating), ha='center', va='bottom')

plt.show()
```
