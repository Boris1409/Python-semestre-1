<h3> 2. Reemplazar los valores nulos, por un valor por defecto en este caso 'cero' (column:rating).
</h3>

```python
datos = pd.read_csv("games.csv", index_col="Id")
datos_copy = datos.copy().sort_values('Rating').head(10)
datos.head(10)
```

```python
datos.isnull().any()
```

```python
datos['Rating'] = datos['Rating'].fillna(0)
```

```python
datos_reverse = datos.sort_values(['Rating'], ascending= True )
datos_reverse
```
